## Sets

```python
set_var = set()
print(set_var)
print(type(set_var))
```

```python
set_var =  {1,2,3,4,5,3}
set_var
```

```python
set_var = {"Avengers","Ironman","Hitman"}
print(set_var)
print(type(set_var))
```

```python
# Python Sets does not support Indexing
set_var[0]
```

```python
set_var.add("Hulk")
set_var
```

```python
set1 = {"Avengers","Ironman","Hitman"}
set2 = {"Avengers","Ironman","Hitman","Hulk"}
set2.difference(set1)
```

```python
set2.intersection(set1)
```

```python
set2
```

```python
set2.intersection_update(set1)
set2
```

```python
set2
```

```python
set2.difference_update(set1)
set2
```

## Dictionaries

```python
dic = {}
```

```python
print(type(dic))
```

```python
d = dict()
print(type(d))
```

```python
my_dict = {
    "Car1":"Audi",
    "Car2":"BMW",
    "Car3":"Mercedez Benz"
}
print(type(my_dict))
```

```python
my_dict
```

```python
for x in my_dict:
    print(x)
```

```python
my_dict["Car1"]
```

```python
for x in my_dict.values():
    print(x)
```

```python
for x in my_dict.items():
    print(x)
```

```python
my_dict['Car3'] = "Rolls Royce"
```

```python
my_dict
```

## Nested Dictionaries

```python
car1_model = {"Mercedez":1960}
car2_model = {"Audi":1970}
car3_model = {"Ambassador":1980}
car_type = {
    "Car1":car1_model,
    "Car2":car2_model,
    "Car3":car3_model
}
car_type
```

```python
car_type['Car1']['Mercedez']
```
